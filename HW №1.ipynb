### 3 task

```python
def diff1(f, x0, h):
    return (f(x0 + h) - f(x0))/h
def diff2(f, x0, h):
    return (f(x0 + h) - f(x0 - h))/(2*h)

hs = np.logspace(-16,0, num = 50)
err1 = []
err2 = []
for h in hs:
    err1.append(abs(diff1(np.sin, 1, h) - np.cos(1)))
    err2.append(abs(diff2(np.sin, 1, h) - np.cos(1)))
    
M2 = M3 = 1

plt.figure(figsize = (18, 10))

plt.loglog(hs, err1, 'r-', ms = 16)
plt.loglog(hs, err2, 'k-', ms = 16)
plt.loglog(hs, hs*M2/2, 'r-', label = "first order", lw = 3)
plt.loglog(hs, hs**2*M3/6, 'k-', label = "first order", lw = 3)
plt.ylim(1e-14, 1)
plt.ylabel('$\Delta$')
plt.xlabel('Step')
plt.grid()
plt.show()
```
